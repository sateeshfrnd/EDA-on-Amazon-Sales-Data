## Data Cleaning

#### Change the data type of columns having numerical values

```python
df = data
```

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'').str.replace(",",'').astype('float64')
```

```python
df['actual_price'] = df['actual_price'].str.replace("₹",'').str.replace(",",'').astype('float64')
```

```python
data['discount_percentage'].unique()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64') / 100
```

```python
data['discount_percentage'].unique()
```

##### Handle rating column

```python
df['rating'].unique()
```

Here rating column having strange value '|', replacing value with median or mean value of the existing ratings as a sensible approach to retain the original distribution.

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['rating'].unique()
```

```python
rate_median_value = df['rating'].median()
rate_median_value
```

```python
# Replace missing values in the 'rating' column with the median rating value
df['rating'].fillna(rate_median_value, inplace=True)
```

```python
df['rating'].unique()
```

```python
# Define rating categories based on numerical ranges
def categorize_rating(rating):
    if rating < 2:
        return 'Poor'
    elif 2 <= rating < 3:
        return 'Fair'
    elif 3 <= rating < 4:
        return 'Good'
    else:
        return 'Excellent'
```

```python
# Apply the categorize_rating function to create a new 'Rating Category' column
df['rating_category'] = df['rating'].apply(categorize_rating)
```

```python
df['rating_category'].unique()
```

```python
df['rating_category'] = df['rating_category'].astype('category')
```

```python
# Reorder categories
df['rating_category'] = df['rating_category'].cat.reorder_categories(['Fair', 'Good', 'Excellent'])
```

```python
df['rating_category'].unique()
```

```python
df[['rating','rating_category']]
```

##### Handle rating_count

```python
df['rating_count'].unique() 
```

```python
# Change the rating_count column Data Type
df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
df['rating_count'].unique() 
```

### 1. What is the average rating for each product category ?

```python
avg_rating_per_category = df.groupby('category')['rating'].mean()
avg_rating_per_category
```

```python
ratings = avg_rating_per_category.sort_values(ascending=False)
ratings
```

```python
# Create the plot
plt.figure(figsize=(14, 6))
barplot = sns.barplot(x=ratings.values, y=ratings.index, palette="viridis")

# Add labels and title
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Product Category', fontsize=14)
plt.title('Average Rating by Product Category', fontsize=16)

# Add data labels with space
for index, value in enumerate(ratings.values):
    barplot.text(value + 0.05, index, f'{value:.2f}', color='black', ha="left", fontsize=12)

# Show the plot
plt.show()
```

**Observations:**
 - *"OfficeProducts"* Category is having more rating compared to other categories, it suggests that products in this category are generally better rated or more positively received by customers.

### 8. What is the correlation between discounted_price and rating?

```python
correlation = df['discounted_price'].corr(df['rating'])
correlation
```

```python
correlation_matrix = df[['discounted_price', 'rating']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Discounted Price vs. Rating')
plt.show()
```

The correlation coefficient between the discounted price and the rating is approximately 0.1203.

Since the correlation coefficient is positive and relatively small (close to zero), it suggests a weak positive linear relationship between the discounted price and the rating. This means that there is a slight tendency for products with higher discounted prices to have slightly higher ratings, but the relationship is not very strong.
